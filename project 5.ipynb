```python
data_inp=data.drop('quality',axis=1)
data_op=data.quality
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
lr=LogisticRegression()
max_score=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=tts(data_inp,data_op,random_state=i,test_size=0.20,stratify=data_op)
    lr1=lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    print('accuracy score at',i,'is',accuracy)
    if accuracy>max_score:
        max_score=accuracy
        final_state=i
print('max accuracy is ',max_score,'at',final_state)
```

```python
x_train,x_test,y_train,y_test=tts(data_inp,data_op,random_state=97,test_size=0.33,stratify=data_op)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
train_pred=lr.predict(x_train)
accuracy=accuracy_score(y_test,y_pred)
accuracy1=accuracy_score(y_train,train_pred)
print(accuracy,accuracy1)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
