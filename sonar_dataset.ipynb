```python
max_score=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=tts(X,Y,random_state=i,test_size=0.33,stratify=Y)
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    train_pred=lr.predict(x_train)
    acc=accuracy_score(y_test,y_pred)
    acc1=accuracy_score(y_train,train_pred)
    print('at random state:',i)
    print('Test accuracy is:',acc)
    print('Train accuracy is:',acc1)
    if acc>max_score:
        max_score=acc
        final_state=i
print('Maximum accuracy observed is:',max_score,'at random state:',final_state)
```
